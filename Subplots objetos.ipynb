# Sunplot con orientacion a objetos

```python
#importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()
```

```python
datos = pd.read_excel(
    "D:/Backup/Documents/Scripts pyton/Graficas_2d_python/datos yopal.xlsx"
    )
datos.head(2)
```

```python
fig, axes = plt.subplots()
axes.scatter(datos['ingreso'], datos['edad'])
plt.show()
```

```python
fig,  axes = plt.subplots(nrows=1, ncols= 2)
axes[0].scatter(datos['ingreso'], datos['edad'])
axes[1].hist(datos['edad'])
plt.show()
```

```python
fig, axes= plt.subplots(2,3)
axes[0,0].hist(datos['edad'])
axes[0,1].hist(datos['ingreso'])
axes[0,2].plot(datos['ingreso'], datos['edad'], 'r')
axes[1,0].boxplot(datos['ingreso'])

fig.tight_layout()
plt.show()
```

```python
# graficas con titulo 

fig, axes= plt.subplots(2,2, figsize=(9,5))

axes[0,0].hist(datos['edad'], label = " edad")
axes[0,0].set_title('Histograma de la edad')
axes[0,0].set_ylabel('Conteo')
axes[0,0].legend()

axes[0,1].plot(datos['ingreso'], datos['edad'], 'r')

axes[1,0].boxplot(datos['ingreso'], vert =False, patch_artist =True, notch =True)

fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize = (5,5))
axes.hist(datos['edad'], label = " edad",color='r')
axes.set_title( 'Título')
axes.set_ylabel('y')
axes.set_xlabel('x')

axes.legend()
plt.show()
```

```python
fig, axs = plt.subplots(4, 2, figsize=(10, 15), constrained_layout=True)
index = []
cols = datos.select_dtypes(include= 'object').columns 
for i in range(4):
    for j in range(2):
        index.append(axs[i][j])
for i in range(len(cols)):
    sns.countplot(x=cols[i], hue='genero', data=datos, ax=index[i]).set(title=cols[i], xlabel=None, ylabel=None)
plt.show()
axs[3, 1].remove()
fig.supxlabel("Value").set_fontsize(14)
fig.supylabel("Count").set_fontsize(14)
fig.suptitle("Distribution of data for the new features").set_fontsize(16)
```
