# Graficas con seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
datos = pd.read_excel(
    "D:\Antonio Medina\Documents\Python Scripts\Graficas_2d_python\datos yopal.xlsx"
    )
datos.head(2)
```

```python
sns.histplot(
    data= datos,
    x='ingreso',
    hue= 'genero')
plt.show()
```

```python
sns.histplot(data= datos, x='ingreso', y= 'edad', hue= 'genero')
plt.show()
```

```python
sns.pairplot(data =datos)
plt.show()
```

```python
datos.columns
```

```python
sns.pairplot(
   data =datos,
   hue='genero'
)
plt.show()
```

```python
sns.heatmap(
    datos.corr(),
    annot= True,
    cmap = "RdBu",
    vmax= 1,
    vmin= -1)
plt.show()
```

```python
sns.lmplot(
    data= datos,
    x= "edad",
    y= "ingreso",
    )
plt.show()
```

```python
sns.lmplot(
    data= datos,
    x= "edad",
    y= "ingreso",    
    hue= "genero",
    height=5
    )
plt.show()
```

```python
sns.histplot(
    data= datos,
    x= 'edad'
)
plt.axvline(
    x= datos.edad.mean(),
    color= 'red',
    linestyle ='dashed',
    linewidth= 2
)
plt.show()
```

```python
# importar libreria 
import seaborn as sns
import matplotlib.pyplot as plt

# colocar estilo a seaborn
sns.set_theme(style="darkgrid")

#importar data
df = sns.load_dataset("penguins")

# Grafica 

sns.displot(
    df, x="flipper_length_mm", col="species", row="sex",
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),
)
plt.show()
```
