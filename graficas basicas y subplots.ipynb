# Graficos basicos y subplots

```python
#importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_yoyal = pd.read_excel(
    "D:\Antonio Medina\Documents\Python Scripts\Graficas_2d_python\datos yopal.xlsx"
    )
df_yoyal.head(2)
```

```python
df_yoyal['ingreso'].plot(
    kind= 'box',
    vert = False,
    xlabel='Valor expresado en millones',
    title= 'INGRESO PEA EN YOPAL CASANARE',
    rot= 90
    )
plt.show()
```

```python
df_yoyal.plot(
    kind= "scatter",
    x= 'ingreso',
    y='edad',
    title= 'Relacón edad ingreso',
    
    )
plt.show()
```

# Graficos orietados a onjetos 

```python
plt.scatter (df_yoyal['ingreso'],df_yoyal['edad'])
plt.xlabel('Ingreso')
plt.title(' Ingresos vs edad')
plt.show()
```

# Subplots 

Es para hacer varios al mismo tiempo 

```python
plt.subplot(1,2,1)
plt.scatter(df_yoyal['ingreso'], df_yoyal['edad'], color ='red')
plt.title('Relación del inglreso versus edad ')
plt.xlabel('Ingreso')
plt.ylabel('Edad')

plt.subplot(1,2,2)
plt.boxplot(df_yoyal['ingreso'])
plt.title('Puntos atípicos del ingreso')
plt.xlabel('Ingreso')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.boxplot(df_yoyal['ingreso'])

plt.subplot(1,2,2)
plt.pie(df_yoyal['experiencia_total'])
plt.title('Experiencia')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.boxplot(df_yoyal['ingreso'], vert= False)

plt.subplot(2,1,2)
plt.pie(df_yoyal['experiencia_total'])
plt.title('Experiencia')
plt.show()
```
