```python
# dos primeros parametros posición, tamaño, 
fig =plt.figure()
axes = fig.add_axes([0.1,0.1,1,0.9])
axes.scatter(datos['ingreso'], datos['edad'])
plt.show()
```

```python
# dos primeros parametros posición, tamaño, 
# Son objetos diferetes, pero en un mismo lienso
fig =plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.5,0.7,0.4,0.3])

axes1.scatter(datos['ingreso'], datos['edad'])
axes2.scatter( datos['edad'],datos['ingreso'])

plt.show()
```

```python
fig =plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.5,0.7,0.4,0.3])

axes1.scatter(datos['ingreso'], datos['edad'])
axes2.hist( datos['edad'])
axes2.set_facecolor('gray')
plt.show()
```
